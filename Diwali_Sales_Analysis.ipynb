## 🌍 Top 10 States by Number of Orders

In this step, we analyze the top 10 states based on the total number of orders.

- Grouped data by State  
- Calculated total number of orders  
- Sorted values in descending order  
- Selected top 10 states  
- Visualized using barplot  

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15, 5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
plt.show()
```

### 📌 Observation

- The top states contributing the highest number of orders include Uttar Pradesh, Maharashtra, and Karnataka  
- These states have the highest customer activity during Diwali sales  
- Order distribution is concentrated in a few key states
